* 방법(5) 함수를 만들어서 사용 - lambda

<font color=red><b>
    df[컬럼].apply(  lambda x컬럼 : myfunc(x컬럼)  )

```python
"""
함수이름 : my_extpnt
파라미터 : prmgen
    res m:1  f:0
return res
"""
def my_extpnt(prmgen):
    res = 0
    if prmgen =='m':
        res = 1
    #elif prmgen =='f':
    else:
        res = 0
    return res    
# my_extpnt('m')   #------1
# my_extpnt('f')   #------1
```

```python
user  = pd.read_csv("./dataset/user.csv")

# x = df['gen']
# my_extpnt( x )
user['extpnt'] = user['gen'].apply(lambda x : my_extpnt(x)  )
user
```

* age_gubun 신규컬럼 생성
<pre>
20 ~ 29  : 2
30 ~ 39  : 3
40 ~ 49  : 4

```python
user  = pd.read_csv("./dataset/user.csv")
def myfunc(age):
    res = 0
    if age in range(20, 30):  #20,21...29
        res = 2
    elif age in range(30, 40):
        res = 3
    else:
        res = 4
    return res
myfunc(22)

user['age_gubun'] = user['age'].apply(lambda x : myfunc(x))
user
# if age>=20 and age<=29:
#     print(2)
# elif age>=30 and age<=39:
#     print(3)
# else:
#     print(4)
```
