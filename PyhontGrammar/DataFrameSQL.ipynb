# MERGE() = db:join
* how : {'left', 'right', 'outer', 'inner'}
* 키가 동일하면 : on
* 왼쪽 오른쪽 프레임에 조인해야 할 키가 다르다 : left_on, right_on

```python
data_list = [   [10,'ACCOUNTING','NEW YORK'],
                [20,'RESEARCH','DALLAS'],
                [30,'SALES','CHICAGO'],
                [40,'OPERATIONS','BOSTON']
             ]
col_list = ['deptno', 'dname', 'loc']
dept = pd.DataFrame(data   = data_list,  #[[]]
                      columns = col_list   #[]
                 )  
print(dept.info())  
dept
```

```python
# ORACLE : select * from emp e, dept d where e.deptno = d.deptno;
# ANSI SQL : select * from emp e INNER JOIN dept d ON e.deptno = d.deptno;
emp = df.drop(  ['new_col', 'ename_job', 'comm2', 'sal+comm2'] , axis=1)
emp_dept = pd.merge(emp, dept, how="inner", on="deptno")
emp_dept
```

* (left  == right) outer join

```python
# ORACLE : select * from emp e, dept d where e.deptno(+) = d.deptno;
# ANSI SQL : select * from emp e RIGHT OUTER JOIN dept d ON e.deptno = d.deptno;
emp = df.drop(  ['new_col', 'ename_job', 'comm2', 'sal+comm2'] , axis=1)
#emp_dept = pd.merge(emp, dept, how="right", on="deptno")
emp_dept = pd.merge(dept, emp, how="left", on="deptno")
emp_dept
```

* 두 테이블 조인 시 연결할 컬럼 이름이 다른 경우 : left_on , right_on
* emp.DNO == dept.deptno

```python
emp = df.drop(  ['new_col', 'ename_job', 'comm2', 'sal+comm2'] , axis=1)
# emp.columns = ['empno', 'ename', 'job', 'mgr', 'hiredate', 'sal', 'comm', 'DNO']
emp = emp.rename(columns={"deptno":"DNO"})
emp
#emp.DNO == dept.deptno
emp_dept = pd.merge(emp, dept, how="inner", left_on="DNO", right_on="deptno")
emp_dept
```
